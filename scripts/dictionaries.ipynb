```python
from collections import defaultdict
import csv

# declare the dictionaries
parties = defaultdict(str) # key = president names, value = political party
# parties = {} # this is a normal dictionary, not a defaultdict
party_count = defaultdict(int) # key = political party, value = number of presidents in that party
presidents = defaultdict(str) # key = president number, value = president name

# Open the presidents.csv
input_file = "../../files/data/presidents.csv"
with open(input_file, "r") as data:
    librarian = csv.reader(data, delimiter=",")
    # we want to skip the header line
    header = next(librarian)
    for line in librarian:
        # print(line,end='')
        # line = line.strip() <- depreciated
            # split each line into a list <- depreciated
            # generally, we don't want to split the line ourselves
            # columns = line.split(",") # there's a module that does this for us

        # load the president dictionary
        # we can also assign names to the different columns
        # presidents[columns[0]] = columns[1] <- depreciated
        # each line is a list now

        # load the parties dictionary
        # parties[columns[1]] = columns[5] <- depreciated
        num = line[0].strip()
        name = line[1].strip()
        presidents[num] = name
        parties[name] = line[5].strip()


        party_count[parties[name]] += 1
```

```python
# let's get stuff from our dictionaries
print(presidents.get('16')) # get the name of the 16th president
print(parties.get(presidents.get('16'))) # get the political party of the 16th president
```

```python
# loop over the dictionaries
for num in presidents.keys():
    print(f"President {num} was {presidents[num]} and was a member of the {parties[presidents[num]]} party.")
```

```python
# instead of printing, let's write out the output
outfile = open("../../files/data/pres.out", "w")

# Another way to loop over a dictionary
for num, name in presidents.items():
    outfile.write(f"President {num} was {name} and was a member of the {parties[name]} party.\n")
# the data doesn't get written out until we close the file
outfile.close()
# This doesn't handle the first line
```
